### Importing the Required Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Reading the dataset 

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.describe()
```

### Exploratory Data Analysis

```python
df.info()
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(26,12))
sns.heatmap(corr,cmap='viridis', annot=True)
```

### Taking a Sample of the Data

```python
data = df.sample(frac=0.25,random_state=1)
```

```python
data.info()
```

### Checking the Spread of the Data

```python
data.hist(figsize=(20,20))
```

```python
data['Class'].value_counts()
```

#### Dropping the column as the values mentioned in the column are inconclusive

```python
x = (data.drop('Class', axis = 1))
y = data['Class']
```

### Building a Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

#### Splitting the data into train and test set.

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
model = LogisticRegression()
```

## Training the Model

```python
model.fit(x_train,y_train)
```

### Checking the quality of the results generated by Algorithm

```python
from sklearn.metrics import classification_report
```

### Accuracy close to 99.88 for the given sample of data

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
prediction = model.predict(x_test)
```

```python
Log_reg = classification_report (y_test,prediction)
```

```python
print(Log_reg)
```
